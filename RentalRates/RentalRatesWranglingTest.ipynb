```python
import pandas as pd
import time
from geopy.geocoders import Nominatim
geolocater = Nominatim(user_agent="suddenjan")
```

```python
rental_rates=pd.read_csv("Resources/fy2024rent.csv", dtype=str)

rental_rates
```

```python
rental_rates['ZIPCode'].head(50)
```

```python

rental_rates=rental_rates.rename(columns={"HUD Metro Fair Market Rent Area Name": "City", "ZIPCode": "Zip Code", "erap_fmr_br0": "Studio", "erap_fmr_br1": "1Bed", "erap_fmr_br2": "2Bed", "erap_fmr_br3": "3Bed", "erap_fmr_br4": "4Bed"})
rental_rates = rental_rates[["City", "Zip Code", "Studio", "1Bed", "2Bed", "3Bed", "4Bed"]]
rental_rates
```

```python
rental_rates['Zip Code'].head(50)
```

```python
rental_rates['Country'] = 'USA'
rental_rates['Zip/Country'] = rental_rates['Zip Code'].astype(str) + ', ' + rental_rates['Country'].astype(str)
rental_rates = rental_rates[["City", "Zip/Country", "Studio", "1Bed", "2Bed", "3Bed", "4Bed"]]
rental_rates
```

```python
locations = rental_rates["Zip/Country"]
locations = [str(location) for location in locations]
```

```python
processed_requests = 0
max_requests_before_pause = 3000
```

```python
latitudes = []
longitudes = []

for location in locations:
    try:
        location_data = geolocater.geocode(location, timeout=None)
        if location_data:
            latitudes.append(location_data[0])
            longitudes.append(location_data[1])
#            print(f"Location: {location}")
 #           print(f"Latitude: {location_data.latitude}, Longitude: {location_data.longitude}")
  #          print()
        else:
            latitudes.append(None)
            longitudes.append(None)
            print(f"Geocoding failed for: {location}")
    except Exception as e:
        print(f"Error occurred for {location}: {e}")

# Add latitude and longitude columns to the DataFrame
rental_rates['Latitude'] = latitudes
rental_rates['Longitude'] = longitudes
```
